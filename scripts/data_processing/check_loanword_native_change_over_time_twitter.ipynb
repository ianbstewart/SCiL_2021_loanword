# Check loanword/native change over time in Twitter
Let's look for really basic examples of loanword and native word adoption rates over time.

```python
import pandas as pd
import json
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import os
import re
import gzip
```

### Query data
Let's define some query phrases for French and Spanish, and then count their frequency over time.

```python
archive_dir = '/hg190/corpora/twitter-crawl/new-archive/'
file_matcher = re.compile('tweets-\w+-\d{2}-\d{2}-\d{2}-\d{2}\.gz')
data_files = [os.path.join(archive_dir, x) for x in os.listdir(archive_dir) if file_matcher.match(x) is not None]
print(data_files)
```

```python
fr_test_phrases = ['tweeter', 'faire un tweet', 'googler', 'chercher sur google']
es_test_phrases = ['tuitear', 'hacer un tweet', 'googlear', 'buscar en google']
fr_test_phrase_matcher = re.compile('|'.join(fr_test_phrases))
es_test_phrase_matcher = re.compile('|'.join(es_test_phrases))
```

```python
fr_phrase_match_data = []
for f in data_files:
    print('processing file %s'%(os.path.basename(f)))
    f_data = []
    for l in gzip.open(f, 'r'):
        try:
            l_data = json.loads(l.strip())
            if(l_data['lang']=='fr'):
                l_txt_match = fr_test_phrase_matcher.search(l_data['status_txt'].lower())
                if(l_txt_match is not None):
                    print('match %s'%(l_txt_match))
                    break
        except Exception as e:
            pass
```
