# Expand loanword light verbs
We have collect n-grams that cooccur with loanword in hopes of finding more light verbs to boost the low coverage. We mined [here](collect_loanword_n_grams.sh), so let's see what we can find.

## Load data

```python
import numpy as np
import pandas as pd
loanword_n_gram_counts = pd.read_csv('../../data/mined_tweets/loanword_n_gram_counts.tsv', sep='\t', index_col=False)
display(loanword_n_gram_counts.head())
```

#### Examine top n-grams

```python
import re
top_k = 50
for loanword_i, data_i in loanword_n_gram_counts.groupby('loanword'):
    print(loanword_i)
    # restrict to n-grams where loanword is not initial
    initial_loanword_matcher = re.compile(f'^{loanword_i}')
    data_i = data_i[data_i.loc[:, 'n_gram'].apply(lambda x: initial_loanword_matcher.search(x) is None)]
    data_i.sort_values('count', inplace=True, ascending=False)
    display(data_i.head(top_k))
```
